```python
import math

print("Let's Approximate pi!")
print('Format: (A * pi) / B')

# Collect user values for approximations 
# Handling Error info retrieved from https://docs.python.org/3/tutorial/errors.html

validChoiceInput = ['y','Y','n','N']
validYes = ['y','Y']
validNo = ['n','N']

# Value A
while True:
    choiceA = input('Do you want to input an integer value A for use in the above format? (y/n):')
    if choiceA not in validChoiceInput:
        print ("Looks like that didn't work. Try inputting 'y' for yes or 'n' for no.")
    elif choiceA in validNo:
        print ("Alrighty.")
        userInputA = 1
        break
    elif choiceA in validYes: 
        while True:
            try:
                userInputA = int(input('Awesome! Please input an integer value A for use in the above format:'))
                break
            except ValueError:
                print("Looks like what you inputted wasn't an integer. Try Again!")
        break
        
# Value B
while True:
    choiceB = input('Do you want to input an integer value B for use in the above format? (y/n):')
    if choiceB not in validChoiceInput:
        print ("Looks like that didn't work. Try inputting 'y' for yes or 'n' for no.")
    elif choiceB in validNo:
        print ("Alrighty.")
        userInputB = 1
        break
    elif choiceB in validYes:
        while True:
            try:
                userInputB = int(input('Awesome! Please input an integer value B for use in the above format:'))
                break
            except ValueError:
                print ("Looks like what you inputted wasn't an integer. Try Again!")
        break

# Collect user value for precision of approximations
while True:
    try:
        userInputP = int(input("Please input an integer for your precision(# of decimal places displayed) when approximating pi:"))
        break
    except ValueError:
        print("Looks like what you inputted wasn't an integer. Try Again!")
precision = '{0:<8.' + str(userInputP) + 'f}'    # For use later: precision.format(pi)

# Review userInput
print('------------------------------------------------')
print('Review of what we have for our pi approximation:')
print('Pi Format: ( A * pi ) / B')
print('A =',userInputA, '| B =', userInputB)
print('Value of Pi: (',userInputA,'* pi ) /',userInputB )
print('Desired Precision:',userInputP, 'decimal places')
print('------------------------------------------------')

# Approximation 1
# Found here: http://functions.wolfram.com/Constants/Pi/06/01/01/0003/
result1 = []
app1 = []
x1 = 0

while True:
    num_1 = (4 * ( ((-1) ** x1) / ((2 * x1) + 1) )) # Approximate pi
    pi_1 = ( (userInputA * num_1) / userInputB ) # Approximate pi with userInputs
    
    result1.append(pi_1)
    app1.append(sum(result1))
    
    if x1 >= 2000:
        break
    
    x1 += 1
    
# Approximation 2
# Madhava Approximation
# Found here: https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Other_classical_formulae
result2 = []
app2 = []
x2 = 0

while True:
    num_2 = (math.sqrt(12) * ( ((-3) ** (-x2) ) / ( (2 * x2) + 1) )) # Approximate pi
    pi_2 = ( (userInputA * num_2) / userInputB ) # Approximate pi with userInputs

    result2.append(pi_2)
    app2.append(sum(result2))

    if x2 >= 2000:
        break
        
    x2 += 1
    
# Approximation 3
# Bailey 1997 Approximation
# Found here: https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Efficient_methods
result3 = []
app3 = []
x3 = 0

while True:
    num_3 = ((1 / (16 ** x3) ) * ( (4 / ((8 * x3) + 1)) - (2 / ((8 * x3) + 4)) - (1 / ((8 * x3) + 5)) - (1 / ((8 * x3) + 6)) )) # Approximate pi
    pi_3 = ( (userInputA * num_3) / userInputB ) # Approximate pi with userInputs
    
    result3.append(pi_3)
    app3.append(sum(result3))
    
    if x3 >= 2000:
        break
    
    x3 += 1
    
# Approximation 4
# Bellard Approximation
# Found here: https://en.wikipedia.org/wiki/Bellard%27s_formula
result4 = []
app4 = []
x4 = 0

while True:
    term0 = (((-1) ** x4) / (2 ** (10 * x4)) )
    term1 = ((2 ** 5) / ((4 * x4) + 1))
    term2 = (1 / ((4 * x4) + 3) )
    term3 = ((2 ** 8) / ((10 * x4) + 1))
    term4 = ((2 ** 6) / ((10 * x4) + 3))
    term5 = ((2 ** 2) / ((10 * x4) + 5))
    term6 = ((2 ** 2) / ((10 * x4) + 7))
    term7 = (1 / ((10 * x4) + 9))
    
    num_4 = ( (1 / (2 ** 6)) * (term0 * ((-term1) - term2 + term3 - term4 - term5 - term6 + term7) ) ) # Approximate pi
    pi_4 = ( (userInputA * num_4) / userInputB ) # Approximate pi with userInputs
    
    result4.append(pi_4)
    app4.append(sum(result4))
    
    if x4 >= 2000:
        break
        
    x4 += 1
    
# Table Output

# Determine Spacing
space = '{:<' + str(userInputP) +'}' # Adjust spacing for dashes (row 2)
dash = '-' * userInputP + (2*'-')
space2 = ('{:<' + str(userInputP + 2) + '}' ) # Adjust spacing for row 1

if userInputP < 6: # Make Room for text in table
    space = '{:<6}'
    space2 = '{:<6}'
    dash = '--------'

print('Count' + '|  ' + space2.format('Approx 1') + '  |  ' + space2.format('Approx 2') + '  |  ' + space2.format('Approx 3') + '  |  ' + space2.format('Approx 4'))
print('-----' + '|  ' + space.format(dash) + '  |  ' + space.format(dash) + '  |  ' + space.format(dash) + '  |  ' + space.format(dash))

for term in range(21): # Display terms 0 - 20
    col_1 = term
    col_2 = app1[term]
    col_3 = app2[term]
    col_4 = app3[term]
    col_5 = app4[term]
    
    print( "{:<4d}".format(col_1), '| ', precision.format(col_2), ' | ', precision.format(col_3), ' | ', precision.format(col_4), ' | ', precision.format(col_5) )

# Last Term of table
print("{:<4d}".format(2000), '| ', precision.format(app1[2000]), ' | ', precision.format(app2[2000]), ' | ', precision.format(app3[2000]), ' | ', precision.format(app4[2000]))

# Plot Output
import matplotlib
import matplotlib.pyplot as pt

term_n = [] # Create lists for plot
app1_plot = []
app2_plot = []
app3_plot = []
app4_plot = []

for n in range(20): # 20 x values
    term_n.append(n)
    app1_plot.append(app1[n])
    app2_plot.append(app2[n])
    app3_plot.append(app3[n])
    app4_plot.append(app4[n])

pt.plot(term_n, app1_plot, label = 'Approximation 1')
pt.plot(term_n, app2_plot, label = 'Approximation 2')
pt.plot(term_n, app3_plot, label = 'Approximation 3')
pt.plot(term_n, app4_plot, label = 'Approximation 4')

# Plot Titles
pt.title(str('Approximations of ' + str(userInputA) + 'pi/' + str(userInputB)) )
pt.ylabel('Approximation Value')
pt.xlabel('Term Number')
pt.legend(loc = 'upper right')
```
